```python
iris.drop(['Id'], axis=1, inplace=True)
iris.head()
```

```python
iris.Species.value_counts()
```

```python
iris.head()
```

```python
X = iris.iloc[:,:-1]
X.head()
```

```python
y = iris.iloc[:,-1]
y
```

```python
# Creating Train and Test datasets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 50, test_size = 0.25)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier

# https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html
```

```python
dtc1 = DecisionTreeClassifier(random_state=1)
dtc1.fit(X_train, y_train)
```

```python
y_pred_test = dtc1.predict(X_test)
y_pred_train = dtc1.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_train,y_pred_train),"\n",accuracy_score(y_test, y_pred_test))
```

```python
dtc2 = DecisionTreeClassifier(random_state=1, criterion='entropy')
dtc2.fit(X_train, y_train)
```

```python
y_pred_test = dtc2.predict(X_test)
y_pred_train = dtc2.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_train,y_pred_train),"\n",accuracy_score(y_test, y_pred_test))
```

```python
dtc3 = DecisionTreeClassifier(random_state=1, criterion='log_loss')
dtc3.fit(X_train, y_train)
```

```python
y_pred_test = dtc3.predict(X_test)
y_pred_train = dtc3.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_train,y_pred_train),"\n",accuracy_score(y_test, y_pred_test))
```

```python
dtc4 = DecisionTreeClassifier(random_state=1, criterion='gini', max_depth=3)
dtc4.fit(X_train, y_train)
```

```python
y_pred_test = dtc4.predict(X_test)
y_pred_train = dtc4.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_train,y_pred_train),"\n",accuracy_score(y_test, y_pred_test))
```
