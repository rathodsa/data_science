```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,6)
x
```

```python
y = x**2
y
```

# 1. Function way of plotting

```python
# plotting
plt.plot(x,y)

# cosmetics
plt.xlabel('this is my x values')
plt.ylabel('this is my y values')
plt.title('CHART TITLE')
```

```python
plt.subplot(2,2,1) #n_rows,n_cols,ith plot
plt.plot(x,y)
plt.xlabel('this is my x values')
plt.ylabel('this is my y values')
plt.title('y~x')

plt.subplot(2,2,2)
plt.plot(y,x)
plt.xlabel('this is my x values')
plt.ylabel('this is my y values')
plt.title('x~y')

plt.subplot(2,2,3)
plt.plot(x,x**3)
plt.xlabel('this is my x values')
plt.ylabel('this is my y values')
plt.title('cubic')

plt.subplot(2,2,4)
plt.plot(x**3,x)
plt.xlabel('this is my x values')
plt.ylabel('this is my y values')
plt.title('ulta cubic')
```

```python
# plotting
plt.plot(x,y,label = 'quad')
plt.plot(x,x**3, label = 'cubic')

# cosmetics
plt.xlabel('this is my x values')
plt.ylabel('this is my y values')
plt.title('CHART TITLE')
plt.legend()
```

# 2. Object way

```python
paper = plt.figure(figsize = (20,7))
ax1 = paper.add_axes([0.1,0.1,0.35,0.8]) #left,bottom,length,height (in terms of 5 of the size)
ax1.plot(x,y)
ax1.set_xlabel('this is my X')
ax1.set_ylabel('this is my y')
ax1.set_title('TITLE')

ax2 = paper.add_axes([0.45,0.1,0.35,0.8]) #left,bottom,length,height (in terms of 5 of the size)
ax2.plot(y,x)
ax2.set_xlabel('this is my y')
ax2.set_ylabel('this is my x')
ax2.set_title('TITLE')
```

```python
paper = plt.figure(figsize = (20,10))
ax1 = paper.add_axes([0.1,0.1,0.8,0.8]) #left,bottom,length,height (in terms of 5 of the size)
ax1.plot(x,y,color = 'green')
ax1.set_xlabel('this is my X')
ax1.set_ylabel('this is my y')
ax1.set_title('TITLE')

ax2 = paper.add_axes([0.2,0.5,0.2,0.2]) #left,bottom,length,height (in terms of 5 of the size)
ax2.plot(y,x, color = 'red')
ax2.set_xlabel('this is my X')
ax2.set_ylabel('this is my y')
ax2.set_title('TITLE')

#paper.savefig('my_plot_on_20220611.png')

#plt.show()
```
