# Q1 - Swap 2 numbers

```python
# Swap Two Numbers 
def swapValues(a,b):
  #print("Value of a {}".format(a))
  #print("Value of b {}".format(b))
  print("Value of A is {}. Value of B is {}".format(a,b))
  
  # Swap Numbers using temporary variable
  temp = a
  a = b
  b = temp

  print("New Value of a {}".format(a))
  print("New Value of b {}".format(b))
```

```python
swapValues(a = 100, b = 20)
```

```python
# Alternate style is to retun the values of a, b 
# Swap Two Numbers 
def swap2Num(a,b):
  print("Value of a {}".format(a))
  print("Value of b {}".format(b))

  # Swap Numbers using temporary variable
  temp = a
  a = b
  b = temp

  return a,b
```

```python
a_swapped,b_swapped = swap2Num(a = 10, b = 40)
print("New Value of a is {}".format(a_swapped))
print("New Value of b is {}".format(b_swapped))
```

```python
def maxValue_for(list_in):
  """ Function finds the max value of the list
  using for loop"""
  max = list_in[0]
  
  for val in list_in:
    if(val > max):
      max = val
  return max
```

```python
max_val = maxValue_for(list_in=[18, 12, 28, 16, 22, 29, 23, 26])
print("Max Value is",max_val)
```

[start:stop:step] giving no field as start and stop indicates default to 0 and string length respectively and “-1” denotes starting from end and stop at the start, hence reversing string.

The slice statement [::-1] means start at the end of the string and end at position 0, move with the step -1, negative one, which means one step backwards.

```python
def strReverse(input_str):
  print("Input string : ",input_str)
  print("Reversed string: ",input_str[::-1])
```

```python
ip_str = input("Enter a string:")
strReverse(ip_str)
```

### String Revision

What is the slice expression that gives every third character of string s, starting with the last character and proceeding backward to the first?

```python
s = "ABCDEFG"
```

```python
s[::-4]
```

```python
print(s[::-3])
print(s[-1::-3])
print(s[:0:-3])
print(s[-1:0:-3])
```

The third index in the slice expression is -3, which indicates every third character stepping backward.

The first and second indices should be -1 (the last character) and 0 (the first character). Either of these can be allowed to default.


> Any of these expressions will produce the correct answer:
>
> * s[::-3]
* s[-1::-3]
* s[:0:-3]
* s[-1:0:-3]

# Question 10: Create a function that returns a combination of two python dictionary.
For example, there's are two dictionaries such as {'a': 100, 'b': 200} and {'x': 300, 'y': 200}.

The returned dictionary will be {'a': 100, 'b': 200, 'x': 300, 'y': 200}.

```python
# Example 
dict1 = {'Jessa': 70, 'Arul': 80, 'Emma': 55} # elective 1
dict2 = {'Kelly': 68, 'Harry': 50, 'Olivia': 66} # Ele 2

# copy second dictionary into first dictionary
dict1.update(dict2)
dict3 = dict1
# printing the updated dictionary
print(dict3)
```

Note: One thing to note here is that if both the dictionaries have a common key then the first dictionary value will be overridden with the second dictionary value.


```python
# Example
dict1 = {'Jessa': 70, 'Arul': 80, 'Emma': 55}
dict2 = {'Kelly': 68, 'Harry': 50, 'Emma': 55}

# join two dictionaries with some common items
dict1.update(dict2)
# printing the updated dictionary
print(dict1)
print(dict1['Emma'])
# Output 66
```

```python
dict1 + dict2
```

```python
def combineDict(dict1=None, dict2=None):
  dict1.update(dict2)
  return dict1
```

```python
d1 = {'a': 100, 'b': 200}
d2 = {'x': 300, 'y': 200}

# Display the final dictionary
print('Final Dictionary: {}'.format(combineDict(dict1=d1, dict2=d2)))
```
