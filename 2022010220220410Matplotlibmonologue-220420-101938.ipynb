```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(1,10)
```

```python
x
```

```python
y = x**2
```

```python
y
```

```python
plt.plot(x,y)

plt.xlabel('My X')
plt.ylabel('My Y')
plt.title('MY CHART')
```

```python
z = x**3
```

```python
x
```

```python
y
```

```python
z
```

```python
plt.subplot(2,1,1) # no of rows, no of columns, ith chart
plt.plot(x,y)
plt.xlabel('My X')
plt.ylabel('SQR')
plt.title('SQR CHART')


plt.subplot(2,1,2) #FUNCTION
plt.plot(x,z)
plt.xlabel('My X')
plt.ylabel('CUBE')
plt.title('CUBE CHART')
```

```python
#OBJECTS

fig = plt.figure()

ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax1.plot(x,y)
ax1.set_title('My Chart from object method')
ax1.set_xlabel('My X')
ax1.set_ylabel('My Y')

ax2 = fig.add_axes([0.2,0.5,0.3,0.3])
ax2.plot(x,z)
ax2.set_title('CUBIC')
ax2.set_xlabel('small X')
ax2.set_ylabel('Cube')
```
